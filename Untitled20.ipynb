```python
import numpy as np
import matplotlib.pyplot as plt
from dataclasses import dataclass
from enum import Enum
import random
from typing import List
from collections import defaultdict

# ---------- Type Definitions ----------
class ThreatType(Enum):
    """Types of aerial threats"""
    CRUISE_MISSILE = "Cruise Missile"       # High value, medium cost
    DRONE_SWARM = "Drone Swarm"             # Low value, very low cost
    FIGHTER_JET = "Fighter Jet"             # Very high value, high cost
    LOITERING_MUNITION = "Loitering Munition" # Medium value, low cost

class DefenseLayer(Enum):
    """Defense layers in each cluster"""
    ELECTRONIC_WARFARE = "Electronic Warfare"  # Low cost, against cheap threats
    LASER = "High-Energy Laser"                 # Low operational cost, against small threats
    SHORT_RANGE = "Short-Range System"          # Medium cost
    MEDIUM_RANGE = "Medium-Range System"        # High cost

class EngagementResult(Enum):
    """Engagement outcomes"""
    DESTROYED = "Destroyed"
    SUPPRESSED = "Suppressed (EW)"
    PENETRATED = "Penetrated"
    PRE_LAUNCH_NEUTRALIZED = "Pre-launch Neutralized"

# ---------- Main Classes ----------
@dataclass
class Threat:
    """Aerial threat class"""
    id: int
    type: ThreatType
    value: float  # Strategic value (1 to 10)
    cost: float   # Production cost for attacker (million USD)
    position: np.ndarray  # Position in simulation space
    speed: float
    target_cluster_id: int  # Target cluster ID
    is_active: bool = True
    
    def move_toward_target(self, clusters):
        """Move threat toward target cluster"""
        if self.is_active and self.target_cluster_id < len(clusters):
            target_pos = clusters[self.target_cluster_id].position
            direction = target_pos - self.position
            distance = np.linalg.norm(direction)
            
            if distance > 0:
                direction = direction / distance
                self.position += direction * min(self.speed, distance)
                
                # If reached target
                if distance <= self.speed:
                    return True
        return False

@dataclass
class DefenseSystem:
    """A defense system"""
    layer: DefenseLayer
    max_range: float
    engagement_prob: float  # Engagement success probability
    cost_per_engagement: float  # Cost per engagement (million USD)
    cooldown: int  # Cooldown time after firing
    current_cooldown: int = 0
    
    def can_engage(self):
        return self.current_cooldown == 0
    
    def engage(self, threat: Threat) -> EngagementResult:
        """Engage threat and return result"""
        if not self.can_engage():
            return EngagementResult.PENETRATED
            
        self.current_cooldown = self.cooldown
        
        # Simulate success probability
        if random.random() < self.engagement_prob:
            # Electronic warfare only suppresses cheap threats
            if self.layer == DefenseLayer.ELECTRONIC_WARFARE:
                if threat.cost < 0.5:  # Cheap drones
                    return EngagementResult.SUPPRESSED
                else:
                    return EngagementResult.PENETRATED
            else:
                return EngagementResult.DESTROYED
        return EngagementResult.PENETRATED
    
    def update_cooldown(self):
        if self.current_cooldown > 0:
            self.current_cooldown -= 1

class DefenseCluster:
    """A RAAD defense cluster"""
    def __init__(self, cluster_id, position, importance):
        self.id = cluster_id
        self.position = position
        self.importance = importance  # Cluster importance (1 to 10)
        self.systems = self._initialize_systems()
        self.destroyed_threats = []
        self.cost_incurred = 0.0
        
    def _initialize_systems(self):
        """Initialize defense systems for this cluster"""
        systems = [
            DefenseSystem(DefenseLayer.ELECTRONIC_WARFARE, 30, 0.8, 0.01, 2),
            DefenseSystem(DefenseLayer.LASER, 20, 0.9, 0.001, 1),
            DefenseSystem(DefenseLayer.SHORT_RANGE, 15, 0.7, 0.5, 3),
            DefenseSystem(DefenseLayer.MEDIUM_RANGE, 40, 0.6, 2.0, 5)
        ]
        return systems
    
    def defend(self, threats: List[Threat]) -> List[Threat]:
        """Defend cluster against list of threats"""
        remaining_threats = []
        
        # Prioritize threats by distance
        threats.sort(key=lambda t: np.linalg.norm(t.position - self.position))
        
        for threat in threats:
            if not threat.is_active:
                continue
                
            engaged = False
            # Try engagement from cheapest to most expensive layer
            for system in sorted(self.systems, key=lambda s: s.cost_per_engagement):
                if system.can_engage():
                    distance = np.linalg.norm(threat.position - self.position)
                    if distance <= system.max_range:
                        result = system.engage(threat)
                        
                        if result in [EngagementResult.DESTROYED, EngagementResult.SUPPRESSED]:
                            threat.is_active = False
                            self.destroyed_threats.append((threat, result, system.layer))
                            self.cost_incurred += system.cost_per_engagement
                            engaged = True
                            break
            
            if not engaged and threat.is_active:
                remaining_threats.append(threat)
        
        # Update system cooldowns
        for system in self.systems:
            system.update_cooldown()
            
        return remaining_threats

class IntelligentC2:
    """RAAD Intelligent Command and Control Center"""
    def __init__(self, clusters):
        self.clusters = clusters
        self.pre_launch_neutralized = 0
        self.intel_budget = 10.0  # Intelligence budget for left-of-launch ops
        
    def left_of_launch_operation(self, threats: List[Threat]) -> List[Threat]:
        """Neutralize threats before launch operations"""
        remaining_threats = []
        
        # Simulate cyber/information operations
        # High-value threats have higher neutralization probability
        for threat in threats:
            # Neutralization probability proportional to threat value and intel budget
            neutralization_prob = min(0.3 * threat.value * (self.intel_budget / 20), 0.7)
            
            if random.random() < neutralization_prob and self.intel_budget > 0:
                threat.is_active = False
                self.pre_launch_neutralized += 1
                self.intel_budget -= threat.value * 0.5  # Operation cost
            else:
                remaining_threats.append(threat)
                
        return remaining_threats

# ---------- Main Simulation ----------
def simulate_raid_attack(num_clusters=3, num_threats=50):
    """Simulate a mass attack against RAAD model"""
    
    # 1. Create defense clusters
    clusters = []
    for i in range(num_clusters):
        position = np.array([random.uniform(0, 100), random.uniform(0, 100)])
        importance = random.uniform(5, 10)
        clusters.append(DefenseCluster(i, position, importance))
    
    # 2. Create intelligent C2 center
    c2 = IntelligentC2(clusters)
    
    # 3. Generate various threats
    threats = []
    threat_types = list(ThreatType)
    
    for i in range(num_threats):
        threat_type = random.choice(threat_types)
        
        # Determine characteristics based on threat type
        if threat_type == ThreatType.DRONE_SWARM:
            value = random.uniform(1, 3)
            cost = random.uniform(0.01, 0.05)  # Very cheap
            speed = 3.0
        elif threat_type == ThreatType.CRUISE_MISSILE:
            value = random.uniform(6, 9)
            cost = random.uniform(1, 3)
            speed = 8.0
        elif threat_type == ThreatType.FIGHTER_JET:
            value = random.uniform(8, 10)
            cost = random.uniform(30, 80)
            speed = 12.0
        else:  # LOITERING_MUNITION
            value = random.uniform(4, 7)
            cost = random.uniform(0.1, 0.5)
            speed = 5.0
        
        # Starting position at region edge
        start_pos = np.array([
            random.choice([-10, 110]),
            random.uniform(0, 100)
        ])
        
        threat = Threat(
            id=i,
            type=threat_type,
            value=value,
            cost=cost,
            position=start_pos,
            speed=speed,
            target_cluster_id=random.randint(0, num_clusters-1)
        )
        threats.append(threat)
    
    # 4. Execute left-of-launch strategy
    print("=== Phase 1: Left-of-Launch Operations ===")
    active_threats = c2.left_of_launch_operation(threats)
    print(f"Threats neutralized before launch: {c2.pre_launch_neutralized}")
    
    # 5. Simulate movement and engagement for 20 time steps
    print(f"\n=== Phase 2: Layered Cluster Defense ===")
    
    for time_step in range(20):
        # Move active threats
        for threat in active_threats[:]:  # Copy list for safe removal
            if threat.is_active:
                reached = threat.move_toward_target(clusters)
                if reached:
                    print(f"Threat {threat.id} ({threat.type.value}) penetrated cluster {threat.target_cluster_id}!")
                    active_threats.remove(threat)
        
        # Each cluster defends
        for cluster in clusters:
            active_threats = cluster.defend(active_threats)
        
        if not active_threats:
            print(f"All threats neutralized at time step {time_step}.")
            break
    
    # 6. Analyze results
    print(f"\n=== Final Results ===")
    
    total_attack_cost = sum(t.cost for t in threats if not t.is_active)
    total_defense_cost = sum(c.cost_incurred for c in clusters) + (10 - c2.intel_budget)
    
    # Collect statistics
    stats = defaultdict(int)
    for cluster in clusters:
        for threat, result, layer in cluster.destroyed_threats:
            stats[f"destroyed_by_{layer.name}"] += 1
            stats[f"cost_attack_{threat.type.name}"] += threat.cost
            stats[f"value_destroyed_{threat.type.name}"] += threat.value
    
    # Calculate cost-benefit ratio
    cost_benefit_ratio = total_attack_cost / total_defense_cost if total_defense_cost > 0 else 0
    
    return {
        'clusters': clusters,
        'c2': c2,
        'initial_threats': len(threats),
        'penetrated': len([t for t in threats if t.is_active]),
        'pre_launch_neutralized': c2.pre_launch_neutralized,
        'total_attack_cost': total_attack_cost,
        'total_defense_cost': total_defense_cost,
        'cost_benefit_ratio': cost_benefit_ratio,
        'stats': dict(stats)
    }

# ---------- Visualization ----------
def visualize_results(results):
    """Visualize simulation results"""
    
    fig, axes = plt.subplots(2, 2, figsize=(14, 10))
    
    # 1. Threat type distribution chart
    ax1 = axes[0, 0]
    threat_types = [t.name for t in ThreatType]
    destroyed_by_type = []
    
    for t_type in threat_types:
        key = f"value_destroyed_{t_type}"
        destroyed_by_type.append(results['stats'].get(key, 0))
    
    ax1.bar(threat_types, destroyed_by_type, color='skyblue')
    ax1.set_title('Strategic Value of Neutralized Threats by Type')
    ax1.set_ylabel('Total Strategic Value')
    ax1.tick_params(axis='x', rotation=45)
    
    # 2. Cost comparison chart
    ax2 = axes[0, 1]
    labels = ['Attack Cost', 'Defense Cost']
    values = [results['total_attack_cost'], results['total_defense_cost']]
    colors = ['lightcoral', 'lightgreen']
    ax2.bar(labels, values, color=colors)
    ax2.set_title('Attack vs Defense Cost Comparison (Million USD)')
    ax2.set_ylabel('Cost')
    
    # Display cost-benefit ratio
    ax2.text(0.5, max(values)*0.9, f'Attack/Defense Cost Ratio: {results["cost_benefit_ratio"]:.2f}',
             ha='center', fontsize=10, bbox=dict(boxstyle="round,pad=0.3", facecolor="yellow"))
    
    # 3. Defense layer effectiveness chart
    ax3 = axes[1, 0]
    layers = [l.name for l in DefenseLayer]
    kills_by_layer = []
    
    for layer in layers:
        key = f"destroyed_by_{layer}"
        kills_by_layer.append(results['stats'].get(key, 0))
    
    ax3.pie(kills_by_layer, labels=layers, autopct='%1.1f%%', startangle=90)
    ax3.set_title('Defense Layer Contribution to Threat Elimination')
    
    # 4. Final threat status chart
    ax4 = axes[1, 1]
    status_labels = ['Pre-launch Neutralized', 'Destroyed in Defense', 'Penetrated']
    status_values = [
        results['pre_launch_neutralized'],
        results['initial_threats'] - results['pre_launch_neutralized'] - results['penetrated'],
        results['penetrated']
    ]
    ax4.bar(status_labels, status_values, color=['gold', 'lightgreen', 'lightcoral'])
    ax4.set_title('Final Threat Status')
    ax4.set_ylabel('Count')
    
    plt.suptitle(f'RAAD Model Simulation | Total Threats: {results["initial_threats"]}', fontsize=14, y=1.02)
    plt.tight_layout()
    plt.show()
    
    # Print summary
    print("\n" + "="*50)
    print("RAAD Model Simulation Summary:")
    print("="*50)
    print(f"Number of defense clusters: {len(results['clusters'])}")
    print(f"Total incoming threats: {results['initial_threats']}")
    print(f"Pre-launch neutralized: {results['pre_launch_neutralized']}")
    print(f"Penetrated to targets: {results['penetrated']}")
    print(f"Attack/Defense cost ratio: {results['cost_benefit_ratio']:.2f}")
    
    if results['cost_benefit_ratio'] < 1:
        print(" Defense was more cost-effective than attack (KED strategy successful)")
    else:
        print("  Defense cost exceeded attack cost")

# Run simulation
if __name__ == "__main__":
    print("Resilient & Adaptive Air Defense (RAAD) Model Simulation")
    print("="*60)
    
    results = simulate_raid_attack(num_clusters=3, num_threats=40)
    visualize_results(results)
```
